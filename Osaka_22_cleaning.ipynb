```python
# removal of columns with excessive Nulls
okn = ok.drop(['No','Layout','Transaction-price(Unit price m^2)','Total floor area(m^2)','Renovation','Transactional factors','Frontage','Purpose of Use','Frontage roadFBreadth(m)','Frontage roadFClassification','Frontage roadFDirection','Land shape','Region'],axis=1)
```

```python
# Column Renaming
okn.rename(columns = {'Transaction period':'Period'}, inplace = True)
```

```python
okn.Period.value_counts()
```

```python
#Column splitting
okn['Period'].str.split('quarter',expand=True).rename({0:'quarter', 1:'year'}, axis = 1)
```

```python
#splitting period column to format it properly
okn[['quarter','year']]=okn['Period'].str.split('quarter',expand=True)
```

```python
# reg[x] to remove a digit from a string in Quarter
okn['quarter']=okn['quarter'].str.extract('(\d)', expand=False)
```

```python
#removal of spaces
okn['year']=okn['year'].str.strip()
```

```python
#combined column
okn['Period'] = okn['year'] + '_q' + okn['quarter']
```

```python
#value check- 2000 or greater is a string therefore is not sorted right
okn['Area(m^2)'].sort_values()
```

```python
okn['Area(m^2)'].unique()
```

```python
#replacing string values based on assumption 
okn['Area(m^2)'].replace('2,000 m^2 or greater.', '2000',inplace=True)
```

```python
#casting columns to perform calculations
okn = okn.astype({'Transaction-price(total)': float,'Area(m^2)': float})
```

```python
okn['Area(m^2)'].sort_values()
```

```python
#checkup
okn['Area(m^2)'].sort_values()
```

```python
#calculated field for price per meter squared
okn['Price_Per_M^2'] = okn['Transaction-price(total)']/okn['Area(m^2)']
```

```python
#overview of prices to determine classifications of properties
okn['Transaction-price(total)'].describe()
```

```python
#Classification of properties function


def property_type_func(row, q1,q3):


    if row['Transaction-price(total)'] <= q1:
           return 'low_end'
    elif row['Transaction-price(total)'] >= q3:
            return 'high_end'
    else:
            return 'mid_end'
```

```python
#Applying function to data
q1 = okn['Transaction-price(total)'].quantile(0.25)
q3 = okn['Transaction-price(total)'].quantile(0.75)

okn['Property_Classification']=okn.apply(property_type_func,args=(q1,q3), axis=1)
```
